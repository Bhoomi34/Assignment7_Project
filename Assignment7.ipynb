```python
import matplotlib.pyplot as plt
Deaths_counts=dataset['Deaths'].value_counts()
plt.bar(Deaths_counts.index,Deaths_counts.values)
plt.xlabel('Deaths')
plt.ylabel('Count')
plt.title('Count of Each Deaths')
# plt.xticks(rotation=45)
plt.show()
```

```python
plt.hist(dataset['Active'],bins=20)
plt.xlabel('Active')
plt.ylabel('Frequency')
plt.title('Distribution of Active ')
plt.show()
```

```python
Recovered_counts=dataset['Recovered'].value_counts()
plt.pie(Recovered_counts,labels=Recovered_counts.index,autopct='%1.1f%%',startangle=90)
plt.title('Distribution of Recovered')
plt.show()
```

```python
import matplotlib.pyplot as plt
Active_counts=dataset['Active'].value_counts()
plt.plot(Active_counts.index,Active_counts.values)
plt.xlabel('Active')
plt.ylabel('Count')
plt.title('Count of items in Each Active')
plt.xticks(rotation=65)
plt.show()
```

```python
import matplotlib.pyplot as plt
Confirmed_counts=dataset['Confirmed'].value_counts()
plt.scatter(Confirmed_counts.index,Confirmed_counts.values)
plt.xlabel('Confirmed')
plt.ylabel('Count')
plt.title('Count of Confirmed')
# plt.xticks(rotation=45)
plt.show()
```
